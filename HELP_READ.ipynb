Attention, you need an internet connection to use this script since gtts uses internet to convert text to speach.


```python
import PyPDF2
#import pyttsx3
import gtts
from gtts import gTTS

# Open the PDF file
with open('/content/JASA2010.pdf', 'rb') as pdf_file:
    pdfreader = PyPDF2.PdfReader(pdf_file)

    full_text = ''
    number_of_pages = len(pdfreader.pages)

    # Extract text from each page
    for page_num in range(number_of_pages):
        text = pdfreader.pages[page_num].extract_text()
        clean_text = text.strip().replace('\n', ' ')
        full_text += clean_text + ' '

    # Save the extracted text to an audio file
    tts=gTTS(full_text)
    tts.save('output.mp3')
from IPython.display import Audio
Audio('output.mp3')
```
